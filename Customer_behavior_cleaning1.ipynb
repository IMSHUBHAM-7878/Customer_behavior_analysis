```python
import pandas as pd
df = pd.read_csv('customer_shopping_behavior.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum() # count missing values
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(" ","_");
```

```python
df.columns
```

```python
df = df.rename(columns = {"purchase_amount_(usd)":"purchase_amount"})
```

```python
df['review_rating'] = df['review_rating'].fillna(df.groupby('category')['review_rating'].transform('mean'))
```

```python
df.info()
```

```python
df['age_group'] = pd.cut(
    df['age'],
    bins = [0,18,35,60,100],
    labels = ['Young','Adult','Mid_Age','Senior']
)
```

```python
df[['age','age_group']]
```

```python
df['frequency_of_purchases'].unique()
```

```python
frequency_mapping = {
    'Fortnightly' : 14,
    'Weekly' : 7,
    'Monthly' : 30,
    'Quarterly' : 90,
    'Bi-Weekly' : 14,
    'Annually' : 365,
    'Every 3 Months' : 90
}
df['purchase_frequncy_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```
